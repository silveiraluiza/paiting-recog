```python
import pandas as pd
import nltk
nltk.download('punkt')
```

Após carregadas as dependências, o arquivo de dados é lido levando em consideração a coluna "idNoticia" como o index.

```python
news = pd.read_csv("./data/noticias_estadao.csv", dtype={'idNoticia':'int'})
news.set_index('idNoticia', inplace = True)
```

A função *"extract_words"* é utilizada para extrair as palavras do texto e cataloga-las(com a ajuda da função *"catalog"* em um dicionário especifico (reverse_index) com um certo id (o id da noticia).

```python
def catalog(dic, word, pageId):
    if word in dic:
        dic[word].add(pageId)
    else:
        dic[word] = set([pageId])
        
def extract_words(reverse_index, text, news_id):
    words = nltk.word_tokenize(text)
    words = set(words)
    for word in words:
        word = word.lower()
        catalog(reverse_index, word, news_id)
```

Agora, iremos percorrer pelas linhas da tabela de noticias para extrairmos e catalogarmos as palavras do titulo de corpo das noticias.

```python
reverse_index = {}

for index, row in news.iterrows():
    extract_words(reverse_index, row['conteudo'], index)
    extract_words(reverse_index, row['titulo'], index)
```

A função de pesquisa recebe os argumentos divididos por espaço, assim iremos tratar a pesquisa a da seguinte forma:  <br /> Se apenas um argumento for passado, basta retornar os ids catalogados para aquele termo (se ele tiver sido catalogado). <br /> Caso contrário, iremos checar o operador: se for "and", iremos realizar uma interseção entre os dois conjuntos de ids encontrados. Caso contrário (sendo um "ou"), iremos realizar uma união entre dois conjuntos de ids. <br /> Para obter o conjunto de ids utilizamos a função getFromCatalog que retorna a lista de ids referente ao termo (que está armazada no dicionário), caso o termo tenha sido encontrado em algum dos textos e um array vazio caso contrário.

```python
def getFromCatalog(term, dic):
    if term in dic:
        return dic[term]
    else:
        return []
    
def search(query):
    query = query.split()
    term1 = query[0].lower()
    search1 = getFromCatalog(term1, reverse_index)
    if len(query) == 1:
        return list(search1)
    
    operator = query[1]
    term2 = query[2].lower()
    search1 = search1
    search2 = getFromCatalog(term2, reverse_index)
    if operator.lower() == 'and':
        result = search1.intersection(search2)
    else:
        result = search1.union(search2)
    return list(result)
```

### Questão Extra 
Para a questão extra, criamos uma função generica para a pesquisa conjuntiva de termos. <br />Primeiro, procuramos a lista de ids para cada termo, bem como a quantidade e colocamos em uma fila de prioridades para facilitar a interseção, onde os termos menos frequentes são pesquisados primeiro. Se apenas um termo é pesquisado, basta retornar a lista de ids já catalogada para o mesmo. Se não, enquanto houver mais de um item na fila, pegamos os dois primeiros itens (os que tem menores frequencia) e realizamos a interseção destes, e colocamos o resultado de volta na fila (pois precisamos fazer a interseção entre todos). <br/> Quando só houve uma pesquisa na fila, temos o resultado final (pois não é necessário mais fazer interseções). E assim retornamos a lista de ids para a consulta.

```python
from queue import *

def union_search(terms):
    terms = terms.split()
    searchs = PriorityQueue()
    for term in terms:
        term = term.lower()
        result = getFromCatalog(term, reverse_index)
        searchs.put((len(result), result))
    if len(terms) == 1:
        return searchs.get()[1]
    while searchs.qsize()> 1:
        smaller_search1 = set(searchs.get()[1])
        smaller_search2 = set(searchs.get()[1])
        partial_result = smaller_search1.intersection(smaller_search2)
        size = len(partial_result)
        searchs.put((size, list(partial_result)))
    return searchs.get()[1]
```

Considerando o diagrama de Venn, podemos conferir o resultado com algumas equivalencias: <br />
A OR B = A + B - A UNIAO B <br />
A AND B <= MIN(A,B)

```python
assert len(search("pt OR dilma")) == len(search("pt")) + len(search("dilma")) - len(search("pt AND dilma"))
assert len(search("pt OR lula")) == len(search("pt")) + len(search("lula")) - len(search("pt AND lula"))
assert len(search("lula OR dilma")) == len(search("lula")) + len(search("dilma")) - len(search("lula AND dilma"))

assert len(search("lula AND dilma")) <= min(len(search("lula")), len(search("dilma")))
assert len(search("pt AND dilma")) <= min(len(search("pt")), len(search("dilma")))
assert len(search("pt AND lula")) <= min(len(search("pt")), len(search("lula")))

assert len(union_search("pt lula dilma")) <= min(len(search("pt")), len(search("lula")),len(search("dilma")))
```
